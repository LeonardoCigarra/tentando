```python
grouped_df = tabelaRenomeada.groupby(['time'])['salario'].mean()

# Classificar os dados agrupados por média de salário
sorted_df = grouped_df.sort_values(ascending=False)

# Exibir a equipe com a média mais alta de salário
print(sorted_df.head())
```
